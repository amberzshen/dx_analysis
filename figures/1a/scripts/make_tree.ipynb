```python
import matplotlib.pyplot as plt
import networkx as nx
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
T = nx.balanced_tree(2, 5)

options = {
    'node_color': 'black',
    'node_size': 100,
    'width': 3,
}

pos = graphviz_layout(T, prog="dot")
nx.draw(T, pos, **options)
plt.show()
```

```python
edge_list = [
    (0, 5),
    (0, 1),
    (1, 6),
    (1, 7),
    (1, 2),
    (2, 8),
    (2, 3),
    (3, 9),
    (3, 4),
    (4, 10),
    (4, 11),
    (4, 12),
]

# Create a graph from the edge list
T = nx.Graph()  # Use nx.DiGraph() if it's a directed graph
T.add_edges_from(edge_list)

options = {
    'node_color': 'black',
    'node_size': 100,
    'width': 3,
}

pos = graphviz_layout(T, prog="dot")
nx.draw(T, pos, **options)
plt.show()
```

```python
edge_list = [
    (0, 5),
    (0, 1),
    (1, 6),
    (1, 7),
    (1, 2),
    (2, 8),
    (2, 3),
    (3, 9),
    (3, 4),
    (4, 10),
    (4, 11),
    (4, 12),
]

pos = {
    5: (0, 0),
    6: (1, 0),
    7: (2, 0),
    8: (3, 0),
    9: (4, 0),
    10: (5, 0),
    11: (6, 0),
    12: (7, 0),
    0: (3.5, 2.5),
    1: (3.5 + 5/7, 2),
    2: (3.5 + 2 * 5/7, 1.5),
    3: (3.5 + 3 * 5/7, 1),
    4: (3.5 + 4 * 5/7, 0.5)
}

T = nx.Graph() 
T.add_edges_from(edge_list)

options = {
    'node_size': 0,
    'width': 10,
    'edge_color': '#93c47d',
}

mutations = [
    (0, 1), 
    (1, 7),
    (1, 2),
    (2, 3),
    (3, 4),
    (4, 11),
    (4, 12),
]

plt.figure(figsize=(20, 15))
nx.draw(T, pos, **options)

colors = [
    '#e69138ff', # medium orange
    '#723a00ff', # brown
    '#b45f06ff', # dark orange
    '#866d54ff', # grey brown
    '#f9cb9cff', # light orange
    '#d5c5b0ff', # beige
    '#e4a929ff', # gold
] 
for i in range(len(mutations)):

    m = mutations[i]
    
    x_pos = (pos[m[0]][0] + pos[m[1]][0]) / 2 
    y_pos = (pos[m[0]][1] + pos[m[1]][1]) / 2
    
    plt.text(x_pos, y_pos, '', fontsize=75, color=colors[i], fontweight='bold', 
             horizontalalignment='center', verticalalignment='center',
             rotation=0,
             )
    
    
plt.savefig('tree.pdf', transparent=True, bbox_inches='tight')
plt.show()
```
