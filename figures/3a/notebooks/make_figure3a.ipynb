```python
import numpy as np
import matplotlib.pyplot as plt

df = df_1e5

colors = ['#93c47d', '#f6b26b', '#56C1FF']

# Compute means and standard deviations
means = [np.mean(df.ts_num_edges), np.mean(df.linarg_nnz), np.mean(df.genotypes_nnz)]
stds = [np.std(df.ts_num_edges), np.std(df.linarg_nnz), np.std(df.genotypes_nnz)]

# Compute ratios of middle bar to the others
ratio_to_first = means[0] / means[1]
ratio_to_last = means[2] / means[1]

# Bar labels and x positions
labels = ['Data-generating ARG \n (lower bound)', 'Inferred linear ARG', 'Sparse genotype matrix \n (upper bound)']
x_pos = np.arange(len(labels))

# Create the plot
plt.figure(figsize=(7.5, 7))
bars = plt.bar(
    x_pos,
    means,
    yerr=stds,
    color=colors,
    width=0.5,
    capsize=2
)

plt.yscale('log')
plt.ylabel('Number of nonzero entries', fontsize=14)
plt.xticks(ticks=x_pos, labels=labels, fontsize=12)
plt.tick_params(axis='y', labelsize=12)
plt.title('N = 100,000', size=16)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# Add ratio labels on first and last bars
plt.text(
    x=x_pos[0],
    y=means[0] * 1.1,
    s=f"{ratio_to_first:.1f}Ã—",
    ha='center',
    va='bottom',
    fontsize=12
)

plt.text(
    x=x_pos[2],
    y=means[2] * 1.1,
    s=f"{ratio_to_last:.1f}Ã—",
    ha='center',
    va='bottom',
    fontsize=12
)

plt.show()
```
