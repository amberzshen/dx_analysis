```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.patches import FancyArrowPatch
```

```python
edge_list = [
    (-1, 0),
    (0, 1),
    (0, 2),
    (1, 3),
    (2, 4),
    (2, 5),
    (3, 6),
    (3, 7),
    (4, 8),
    (4, 9),
    (5, 10),
]

pos = {
    -1: (0, 4),
    0: (0, 3),
    1: (-2/3, 2),
    2: (2/3, 2),
    3: (-4/3, 1),
    4: (0, 1),
    5: (4/3, 1),
    6: (-2, 0),
    7: (-1, 0),
    8: (0, 0),
    9: (1, 0),
    10: (2, 0),
}

options = {
    'node_size': 2000,
    'width': 10,
    'node_color': '#666666', # light grey
    'edge_color': '#666666', # light grey
    'arrows': True,
    'arrowstyle': '->',
    'arrowsize': 50
}

T = nx.Graph() 
T.add_edges_from(edge_list)
plt.figure(figsize=(15, 15))
nx.draw(T, pos, **options)

mutations = [
    (4, 8),
    (3, 6)
]

colors = [
    '#F77F00', # orange
    '#F77F00', # orange
] 

for i in range(len(mutations)):

    m = mutations[i]

    x_pos = (pos[m[0]][0] + pos[m[1]][0]) / 2 
    y_pos = (pos[m[0]][1] + pos[m[1]][1]) / 2

    plt.text(x_pos, y_pos, '', fontsize=100, color=colors[i], fontweight='bold', 
                horizontalalignment='center', verticalalignment='center',
                rotation=0,
                )
    
plt.show()



```
