```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
edge_list = [
    (11, 8),
    (11, 9),
    (7, 4),
    (7, 5),
    (8, 5),
    (9, 6),
    (10, 6),
    (4, 0),
    (5, 0),
    (5, 1),
    (6, 1),
    (6, 2),
    (6, 3),
    (11, 1),
    (7, 0)
]

pos = {
    0: (1, 0),
    1: (3, 0),
    2: (5, 0),
    3: (7, 0),
    4: (0, 1),
    5: (2, 1),
    6: (5, 1),
    7: (1, 2),
    8: (3, 2),
    9: (5, 2),
    10: (7, 2),
    11: (4, 3),
}

G = nx.DiGraph() 
G.add_edges_from(edge_list)

red_edges = [(7, 0), (11, 1)]
edge_colors = ['#cc0000' if x in red_edges else '#666666' for x in G.edges()]

hollow_nodes = [5, 6]
node_colors = ['white' if n in hollow_nodes else '#666666' for n in G.nodes()]
node_outline_colors = ['#666666' for _ in G.nodes()]
linewidths = [10 if n in hollow_nodes else 0 for n in G.nodes()]



options = {
    'node_size': 2000,
    'width': 10,
    'node_color': node_colors, # light grey
    'edgecolors': node_outline_colors,
    'edge_color': edge_colors, # light grey
    'linewidths': linewidths,
    'arrows': True,
    'arrowstyle': '->',
    'arrowsize': 50
}

plt.figure(figsize=(15, 10))
nx.draw(G, pos, **options)

plt.show()



```
