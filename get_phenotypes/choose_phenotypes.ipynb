```python
import re

def add_space_pattern(s):
    # Add a space before the uppercase letter if surrounded by lowercase letters
    return re.sub(r'(?<=[a-z])(?=[A-Z][a-z])', ' ', s)

with open("price_sumstats.txt", "r") as f:
    sumstats = f.readlines()

keywords = []
with open("price_sumstats.txt", "r") as f:
    for line in f:
        keywords.extend(re.split(r'[._]', line.strip()))
keywords = set(keywords)

keywords = [s for s in keywords if any(c.isupper() for c in s)]
keywords = [add_space_pattern(s) for s in keywords] # not perfect but hopefully good enough

for keyword in keywords:
    print(f'----- {keyword} -----')
    df_filt = df[df["title"].str.contains(rf'\b{re.escape(keyword)}\b', case=False, na=False, regex=True)]
    for t in df_filt.title:
        print(t)
    print()
```
