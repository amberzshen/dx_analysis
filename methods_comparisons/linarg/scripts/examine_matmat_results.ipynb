```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../matmat_benchmark_results.csv')
df
```

```python
plt.scatter(df[df.dot_product_type=='right'].time, df[df.dot_product_type=='left'].time)
plt.xlabel('right dot product time')
plt.ylabel('left dot product time')
plt.show()
```

```python
df32 = df[(df.data_type=='float32') & (df.dot_product_type=='right')]

for method in set(df32.method):
    df32_m = df32[df32.method==method]
    plt.scatter(np.log10(df32_m.n), np.log10(df32_m.time), label=method)
plt.xlabel('number of vectors (log10)')
plt.ylabel('time to compute matmat (log10)')
plt.title('float32')
plt.legend()
plt.show()
```

```python
df64 = df[(df.data_type=='float64') & (df.dot_product_type=='right')]

for method in set(df64.method):
    df64_m = df64[df64.method==method]
    plt.scatter(np.log10(df64_m.n), np.log10(df64_m.time), label=method)
plt.xlabel('number of vectors (log10)')
plt.ylabel('time to compute matmat (log10)')
plt.title('float64')
plt.legend()
plt.show()
```
